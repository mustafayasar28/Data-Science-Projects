## Implementing ANN for linear regression (No Hidden Layer) 

```python
# Sum of squared errors function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_der(x):
    return sigmoid(x) * (1 - sigmoid(x))

def SSE(y, y_hat):
    return 0.50*np.sum((y - y_hat)**2)

def dE_dW1 (y, y_hat,x_1):
    return np.sum(-(y - y_hat) * sigmoid_der(y_hat) * x_1)

def dE_dW0 (y, y_hat):
    return np.sum(-(y - y_hat) * sigmoid_der(y_hat))
```

```python
std_x_train = (x_train - np.mean(x_train)) / (np.std(x_train))
std_y_train =  (y_train - np.mean(y_train)) / (np.std(y_train))

std_x_test  = (x_test - np.mean(x_test)) / (np.std(x_test))
std_y_test = (y_test - np.mean(y_test)) / (np.std(y_test))

# Initialize weights randomly
w_1 = np.random.uniform(-1,1,1)

lr = 0.001
epoch = 500
best_error = 1000000
best_epoch = 0
for e in range(epoch + 1):
    v = w_1 * std_x_train
    y_pred = sigmoid(v)
    error = SSE(std_y_train, y_pred)
    
    if error < best_error:
        best_error = error
        best_epoch = e
        
    w1_derivative = dE_dW1(std_y_train, y_pred, std_x_train)

    w_1 = w_1 - lr * w1_derivative
            
    if e % 100 == 0:
        plt.scatter(std_x_test, std_y_test, c='r', alpha=0.5)
        plt.plot(std_x_test, w_1 * std_x_test , '-b', label="SSE= %.2f" %error + '\n' + "Epoch Number= %d" %(e))
        plt.xlabel('x (Our Data)')
        plt.ylabel('y_hat (Network Outputs)')
        plt.legend()
        plt.grid()
        plt.show()
        
print("Best epoch is:", best_epoch)
print("Best error is:", best_error)
```

## Implementing ANN for Linear Regression (1 Hidden Layer) (Standardized Dataset)

```python
np.random.seed(0)

# Normalize the dataset (min-max normalization)
std_x_train = (x_train - x_train.min()) / (x_train.max() - x_train.min())
std_y_train = (y_train - y_train.min()) / (y_train.max() - y_train.min())

std_x_test = (x_test - x_test.min()) / (x_test.max() - x_test.min())
std_y_test = (y_test - y_test.min()) / (y_test.max() - y_test.min())

std_x_train = std_x_train.reshape(60, 1)
std_x_test = std_x_test.reshape(41, 1)

std_y_train = std_y_train.reshape(60, 1)
std_y_test = std_y_test.reshape(41, 1)

hidden_number_units = [i for i in range(2, 51)]
learning_rates = [0.5, 0.1, 0.01, 0.001, 0.0001]

smallest_error = 1000000
smallest_error_hnu = 0
smallest_error_lr = 0

for hidden in hidden_number_units:
    initial_weights = np.random.rand(1, hidden) # Shape = (1, hidden)
    hidden_weights = np.random.rand(hidden, 1) # Shape = (hidden, 1)
    for lr in learning_rates:
        for epoch in range(2001):
            # First, apply feedforward algorithm
            hidden_input = np.dot(std_x_train, initial_weights) # Shape = (60, hidden)
            assert(hidden_input.shape == (60, hidden))

            hidden_output = sigmoid(hidden_input) # Shape = (60, hidden)
            assert(hidden_output.shape == (60, hidden))

            out_input = np.dot(hidden_output, hidden_weights) # Shape = (60, 1)
            assert(out_input.shape == (60, 1))

            output = sigmoid(out_input) # Shape = (60, 1)
            assert(output.shape == (60, 1))

            # Calculate sum of squared error
            error = SSE(std_y_train, output)
            
            if error < smallest_error:
                smallest_error = error
                smallest_error_hnu = hidden
                smallest_error_lr = lr
                best_hidden_weights = hidden_weights
                best_initial_weights = initial_weights

            # Apply Back Propagation Algorithm for the hidden weights
            derror_doutput = output - std_y_train # Shape = (60, 1)

            doutput_dout_input = sigmoid_der(out_input) # Shape = (60, 1)
            dout_input_hidden_weights = hidden_output # Shape = (60, hidden)

            # Since the result must have the shape (hidden, 1)
            derror_dhiddenweights = np.dot(dout_input_hidden_weights.T, derror_doutput * doutput_dout_input) 

            # Apply back propagation algorithm for the initial weights

            # The following two are already calculated
            # derror_douput shape (60, 1)
            # doutput_dout_input shape (60, 1)
            d_mult = derror_doutput * doutput_dout_input # shape (60, 1)

            dout_input_dhidden_output = hidden_weights # shape (4, 1)

            d_mult2 = np.dot(d_mult, dout_input_dhidden_output.T) # dot product of shapes ( 60, 1 ), (1, hidden)
                                                                    # shape = (60, hidden)

            dhidden_output_dhidden_input = sigmoid_der(hidden_input) # Shape (60, hidden)
            dhidden_input_dinitial_weights = std_x_train

            d_cost_initials = np.dot( dhidden_input_dinitial_weights.T, dhidden_output_dhidden_input * d_mult2 )

            # Update the weights
            hidden_weights -= lr * derror_dhiddenweights
            initial_weights -= lr * d_cost_initials


print("Smallest error: ", smallest_error)
print("Number of units in hidden layer: ", smallest_error_hnu)
print("Learning rate", smallest_error_lr)
```

```python
# Make predictions by using the test set

def predict(predict_set, initial_weights, hidden_weights):
    hidden_input = np.dot(predict_set, initial_weights)
    hidden_output = sigmoid(hidden_input) 
    out_input = np.dot(hidden_output, hidden_weights) 
    predictions = sigmoid(out_input) 
    
    return predictions

# While making predictions, feedforwarding is sufficient
predictions = predict(std_x_test, best_initial_weights, best_hidden_weights)
error = SSE(std_y_test, predictions)

# Make predictions by using the data points uniformly selected in the range 0 and 1
random_points = np.random.random((100,1))
random_points = np.sort(random_points)

rand_predictions = predict(random_points, best_initial_weights, best_hidden_weights)
```

## Implementing ANN for Linear Regression (1 Hidden Layer) (Non-Standardized Dataset)

```python
x_train = x_train.reshape(60, 1)
x_test = x_test.reshape(41, 1)

y_train = y_train.reshape(60, 1)
y_test = y_test.reshape(41, 1)

lr = 0.001

random.seed(19)
np.random.seed(19)

smallest_error = 1000000
smallest_error_hnu = 0
smallest_error_lr = 0

hiddens = [2, 4, 8, 16, 32]

for hidden in hiddens:
    initial_weights = (np.random.rand(len(x_train[0]), hidden))
    hidden_weights = (np.random.rand(hidden, 1))
    error_out = 0
    b1 = (np.random.rand()) * 6 - 3
    b2 = (np.random.rand()) * 6 - 3
    for epoch in range(2001):
        selection = np.arange(0, len(x_train), 1)
        error_out = 0

        # Apply the stochastic learning algorithm 
        while len(selection) > 0: 
            index = random.choice(selection)

            # Remove the selected row
            selection = np.delete(selection, np.argwhere(selection == index))

            # First, apply feedforward algorithm
            Z = np.dot(x_train[index], initial_weights)
            H = sigmoid(Z + b1)
            out_input = np.dot(H, hidden_weights)
            output = out_input + b2

            # Second, apply the back propagation algorithm

            # Apply the backpropagation algorithm for the hidden weights 
            error_out += (np.power((output - y_train[index]), 2))
            
            dcost_doutput = 2 * (output - y_train[index])
            doutput_dout_input = 1
            dout_input_dwo = H
            dcost_wo = dout_input_dwo * (dcost_doutput * doutput_dout_input)

            # Apply the backpropagation algorithm for the initial weights
            dcost_dout_input = dcost_doutput * doutput_dout_input
            dout_input_dah = hidden_weights
            dcost_dah = np.dot(dcost_dout_input, dout_input_dah.T)
            dah_dzh = sigmoid_der(Z)
            dzh_dwi = x_train[index]
            dcost_wi = dzh_dwi * dah_dzh * dcost_dah

            # Update the weights and the bias
            b1 -= lr * (-2 * H.sum() / len(x_train))
            b2 -= lr * (-2 * output.sum() / len(x_train))
            initial_weights -= lr * dcost_wi
            hidden_weights -= (lr * dcost_wo).reshape(hidden, 1)
            
        if error_out[0] < smallest_error:
            smallest_error = error_out[0]
            smallest_error_hnu = hidden
            smallest_error_lr = lr
            best_hidden_weights = hidden_weights
            best_initial_weights = initial_weights
            best_b1 = b1
            best_b2 = b2


print("Smallest error: ", smallest_error)
print("Number of units in hidden layer: ", smallest_error_hnu)
print("Learning rate", smallest_error_lr)
```

```python
Z = np.dot(x_test, best_initial_weights)
H = sigmoid(Z + best_b1)
out = np.dot(H, best_hidden_weights)
prediction = out + best_b2
```

```python
hidden_units = [2, 4, 8, 16, 32]
learning_rates = [0.5, 0.1, 0.01, 0.001, 0.0001]

hidden_weights_for_ann = []
initial_weights_for_ann = []

for hu in hidden_units:
    smallest_error = 100000000
    for lr in learning_rates:
        initial_weights = np.random.rand(1, hu) # Shape = (1, hidden)
        hidden_weights = np.random.rand(hu, 1) # Shape = (hidden, 1)
        for epoch in range(2001):
            # First, apply feedforward algorithm
            hidden_input = np.dot(std_x_train, initial_weights) # Shape = (60, hidden)
            hidden_output = sigmoid(hidden_input) # Shape = (60, hidden)
            out_input = np.dot(hidden_output, hidden_weights) # Shape = (60, 1)
            output = sigmoid(out_input) # Shape = (60, 1)

            # Calculate sum of squared error
            error = SSE(std_y_train, output)
            
            if error < smallest_error:
                smallest_error = error
                smallest_error_lr = lr
                best_hidden_weights = hidden_weights
                best_initial_weights = initial_weights

            # Apply Back Propagation Algorithm for the hidden weights
            derror_doutput = output - std_y_train # Shape = (60, 1)

            doutput_dout_input = sigmoid_der(out_input) # Shape = (60, 1)
            dout_input_hidden_weights = hidden_output # Shape = (60, hidden)

            # Since the result must have the shape (4, 1)
            derror_dhiddenweights = np.dot(dout_input_hidden_weights.T, derror_doutput * doutput_dout_input) 

            # Apply back propagation algorithm for the initial weights

            # The following two are already calculated
            # derror_douput shape (60, 1)
            # doutput_dout_input shape (60, 1)
            d_mult = derror_doutput * doutput_dout_input # shape (60, 1)

            dout_input_dhidden_output = hidden_weights # shape (4, 1)

            d_mult2 = np.dot(d_mult, dout_input_dhidden_output.T) # dot product of shapes ( 60, 1 ), (1, 4)
                                                                    # shape = (60, 4)

            dhidden_output_dhidden_input = sigmoid_der(hidden_input) # Shape (60, 4)
            dhidden_input_dinitial_weights = std_x_train

            d_cost_initials = np.dot( dhidden_input_dinitial_weights.T, dhidden_output_dhidden_input * d_mult2 )

            # Update the weights
            hidden_weights -= lr * derror_dhiddenweights
            initial_weights -= lr * d_cost_initials
            
    hidden_weights_for_ann.append(best_hidden_weights)
    initial_weights_for_ann.append(best_initial_weights)
    print("When number of hidden units is ", hu)
    print("The best learning rate is: ", smallest_error_lr)
    print("The smallest SSE: ", smallest_error)
    print()
```
