## 1.3 Projecting the data onto subspaces and Gaussian Classifier

```python
digits = scipy.io.loadmat('digits.mat')

features = digits['digits']
labels = digits['labels']

y = []

for i in labels:
    y.append(i[0])

labels = np.array(y)

x_train = features[: 2500]
y_train = labels[: 2500]

x_test = features[2500:]
y_test = labels[2500:]
```

```python
def seperate_training_set(reduced, training_y):
    zero_features = []
    one_features = []
    two_features = []
    three_features = []
    four_features = []
    five_features = []
    six_features = []
    seven_features = []
    eight_features = []
    nine_features = []

    for i, element in enumerate(reduced):
        if (training_y[i] == 0):
            zero_features.append(element)
        elif (training_y[i] == 1):
            one_features.append(element)
        elif (training_y[i] == 2):
            two_features.append(element)
        elif (training_y[i] == 3):
            three_features.append(element)
        elif (training_y[i] == 4):
            four_features.append(element)
        elif (training_y[i] == 5):
            five_features.append(element)
        elif (training_y[i] == 6):
            six_features.append(element)
        elif (training_y[i] == 7):
            seven_features.append(element)
        elif (training_y[i] == 8):
            eight_features.append(element)
        elif (training_y[i] == 9):
            nine_features.append(element)

    zero_features = np.array(zero_features)
    one_features = np.array(one_features)
    two_features = np.array(two_features)
    three_features = np.array(three_features)
    four_features = np.array(four_features)
    five_features = np.array(five_features)
    six_features = np.array(six_features)
    seven_features = np.array(seven_features)
    eight_features = np.array(eight_features)
    nine_features = np.array(nine_features)

    # MLEs of the mean vectors
    mean_0 = np.sum(zero_features, axis=0)
    mean_0 = mean_0 / zero_features.shape[0]
    mean_1 = np.sum(one_features, axis=0)
    mean_1 = mean_1 / one_features.shape[0]
    mean_2 = np.sum(two_features, axis=0)
    mean_2 = mean_2 / two_features.shape[0]
    mean_3 = np.sum(three_features, axis=0)
    mean_3 = mean_3 / three_features.shape[0]
    mean_4 = np.sum(four_features, axis=0)
    mean_4 = mean_4 / four_features.shape[0]
    mean_5 = np.sum(five_features, axis=0)
    mean_5 = mean_5 / five_features.shape[0]
    mean_6 = np.sum(six_features, axis=0)
    mean_6 = mean_6 / six_features.shape[0]
    mean_7 = np.sum(seven_features, axis=0)
    mean_7 = mean_7 / seven_features.shape[0]
    mean_8 = np.sum(eight_features, axis=0)
    mean_8 = mean_8 / eight_features.shape[0]
    mean_9 = np.sum(nine_features, axis=0)
    mean_9 = mean_9 / nine_features.shape[0]

    mean_ests = [mean_0, mean_1, mean_2, mean_3, mean_4, mean_5, mean_6, mean_7, mean_8, mean_9]

    # Covariance Matrices
    cov_0 = np.cov(zero_features.T)
    cov_1 = np.cov(one_features.T)
    cov_2 = np.cov(two_features.T)
    cov_3 = np.cov(three_features.T)
    cov_4 = np.cov(four_features.T)
    cov_5 = np.cov(five_features.T)
    cov_6 = np.cov(six_features.T)
    cov_7 = np.cov(seven_features.T)
    cov_8 = np.cov(eight_features.T)
    cov_9 = np.cov(nine_features.T)
    
    cov_ests = [cov_0, cov_1, cov_2, cov_3, cov_4, cov_5, cov_6, cov_7, cov_8, cov_9]
    
    return mean_ests, cov_ests
```

```python
subspace_dims = []
i = 105
while True:
    subspace_dims.append(i)
    i += 5
    if i == 205:
        break
```

## 1.4

```python
classification_errors = []
train_class_errors = []
for dim in subspace_dims:
    pca = PCA(n_components=dim)
    x_train_reduced = pca.fit_transform(x_train)
    x_test_reduced = pca.transform(x_test)
    
    mean_ests, cov_ests = seperate_training_set(x_train_reduced, y_train)
    
    test_correct_count = 0
    # Make Prediction with Gaussian Densities 
    for i, x in enumerate(x_test_reduced):
        accuracies = []
        for c in range(10):
            # For each training row, we have to calculate
            # probability for each class
            y = multivariate_normal.pdf(x, mean=mean_ests[c], cov=cov_ests[c])
            accuracies.append(y)
        index = np.where(accuracies == np.amax(accuracies))
        if (index[0][0] == y_test[i]):
            test_correct_count += 1
    
    acc = test_correct_count / y_test.shape[0] * 100
    print("(Test) Accuracy when subspace dimension is ", dim, ":", acc)
    classification_errors.append(100 - acc)
    
    train_correct_count = 0
    for i, x in enumerate(x_train_reduced):
        accuracies = []
        for c in range(10):
            y = multivariate_normal.pdf(x, mean=mean_ests[c], cov=cov_ests[c])
            accuracies.append(y)
        index = np.where(accuracies == np.amax(accuracies))
        if (index[0][0] == y_train[i]):
            train_correct_count += 1
    
    acc = train_correct_count / y_test.shape[0] * 100
    print("(Training) Accuracy when subspace dimension is ", dim, ":", acc)
    train_class_errors.append(100 - acc)

fig = plt.figure(figsize=(16, 6))

plt.subplot(1,2,1)
plt.plot(subspace_dims, classification_errors)
plt.xlabel('Number of components')
plt.ylabel('Classification Errors on Test Set')

plt.subplot(1,2,2)
plt.plot(subspace_dims, train_class_errors)
plt.xlabel('Number of components')
plt.ylabel("Classification Errors on Training Set")

plt.show()
```

## 2.2

```python
digits = scipy.io.loadmat('digits.mat')

features = digits['digits']
labels = digits['labels']

y = []

for i in labels:
    y.append(i[0])

labels = np.array(y)

x_train = features[: 2500]
y_train = labels[: 2500]

x_test = features[2500:]
y_test = labels[2500:]
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

classification_errors = []
training_classification_errors = []
for comp in range(1, 10):
    lda = LDA(n_components=comp)

    # Project the data using the transformation matrix
    x_train_r2 = lda.fit_transform(x_train, y_train)
    x_test_r2 = lda.transform(x_test)
    
    mean_ests, cov_ests = seperate_training_set(x_train_r2, y_train)

    test_correct_count = 0
    # Make Prediction with Gaussian Densities 
    for i, x in enumerate(x_test_r2):
        accuracies = []
        for c in range(10):
            # For each training row, we have to calculate
            # probability for each class
            y = multivariate_normal.pdf(x, mean=mean_ests[c], cov=cov_ests[c])
            accuracies.append(y)
        index = np.where(accuracies == np.amax(accuracies))
        if (index[0][0] == y_test[i]):
            test_correct_count += 1
    
    acc = test_correct_count / y_test.shape[0] * 100
    print("(Test) Accuracy when subspace dimension is ", comp, ":", acc)
    classification_errors.append(100 - acc)
    
    train_correct_count = 0
    for i, x in enumerate(x_train_r2):
        accuracies = []
        for c in range(10):
            y = multivariate_normal.pdf(x, mean=mean_ests[c], cov=cov_ests[c])
            accuracies.append(y)
        index = np.where(accuracies == np.amax(accuracies))
        if (index[0][0] == y_train[i]):
            train_correct_count += 1
    
    acc = train_correct_count / y_test.shape[0] * 100
    print("(Training) Accuracy when subspace dimension is ", comp, ":", acc)
    training_classification_errors.append(100 - acc)
```

As can be seen, generally the accuracy increases as the LDA dimension increases because we are using more components. However, when the dimension is greater than 7, as can be observed above, the accuracy of tests decreases even though the dimension increases. This is not unexpected because LDA does not guarantee the fact that higher dimension will provide higher accuracy. Moreover, it's likely that this is because of overfitting to the training data. As the model overfits, the bias decreases, however, variance increases. In this case, the training accuracy increases whereas the test accuracy decreases. (Bias-Variance Trade-off)
