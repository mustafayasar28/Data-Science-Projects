```python
# Imports

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans;
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
```

# Question 1

```python
# Read csv file
data = pd.read_csv("falldetection_dataset.csv", header=None)

# The first column does not provide statistically important information
# therefore, drop that column
del data[0]

# Extract the labels (responses) from the dataset 
labels = data.pop(1)

# Convert labels into binary (F = 1, NF = 0)
fall_situation = {'F': 1,'NF': 0}
labels = [fall_situation[item] for item in labels]

labels = np.array(labels, dtype=int)
```

```python
# The features are stored in 'data' data frame
# The labels are stored in 'labels' numpy array

# Perform PCA analysis for the best two principal components

# In order to apply PCA, the dataset must be standardized
data = StandardScaler().fit_transform(data)

# Create PCA object
pca = PCA(n_components=2)

# Fit the pca with features
principalComponents = pca.fit_transform(data)

# Print the variance that the best two principal components capture
exp_var_rat_sum = np.cumsum(pca.explained_variance_ratio_)
print("Variance captured by the best two principal components: ", exp_var_rat_sum[1] * 100, "%")

# Get the principal components
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
# Project the data samples onto the first and second principal components
projected_data = pca.transform(data)
```

```python
projected_data = pd.DataFrame(projected_data,columns=['First_Principal_Component','Second_Principal_Component'])
projected_data.plot(x ='First_Principal_Component', y='Second_Principal_Component', kind = 'scatter')
```

```python
# Run K-means clustering to separate the data into clusters

# Try different number of clusters (N)
N = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for n in N:
    kmeans = KMeans(n)
    kmeans.fit(projected_data)
    
    if n == 2:
        clusters_with_2 = kmeans.fit_predict(projected_data)
    
    identified_clusters = kmeans.fit_predict(projected_data)
    data_with_clusters = projected_data.copy()
    data_with_clusters['Clusters'] = identified_clusters 
    plt.figure()
    plt.scatter(data_with_clusters['First_Principal_Component'],data_with_clusters['Second_Principal_Component'],c=data_with_clusters['Clusters'],cmap='rainbow')
    plt.title("Clustering when the number of clusters is %d" % n)
    plt.show()
    
    
```

```python
# When N = 2 in K-Means, check the degree of percentage overlap/consistency 
# between the cluster memberships and the action labels originally provided

correct_preds = 0
for i in range(566):
    if clusters_with_2[i] == labels[i]:
        correct_preds += 1
        
accuracy = correct_preds / 566 * 100

print("Degree of percentage consistency between the cluster memberships and original labels: ", accuracy)
```
